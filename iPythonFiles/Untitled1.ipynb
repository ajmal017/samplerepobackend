```python
googlenews=GoogleNews(start='05/01/2020',end='05/31/2020')
googlenews.search('Titan')

for i in range(2,3):
    googlenews.getpage(i)
    result=googlenews.result(sort=False)
    dfs=pd.DataFrame(result)
```
